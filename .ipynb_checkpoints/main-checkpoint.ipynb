```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
from urllib.parse import urljoin
```

```python
BASE_URL = 'https://flipkart.com'
```

```python
keyword = input('Enter any product : ')
# /search?q=oppo+earbuds
keyword = keyword.replace(" ", "+")

search_param = f'/search?q={keyword}&sort=popularity'

PRODUCTS_URL = urljoin(BASE_URL, search_param)
print(PRODUCTS_URL)
```

```python
df = pd.DataFrame(columns=['title', 'description', 'reviews', 'ratings'])
```

```python
req = requests.get(PRODUCTS_URL)
soup = bs(req.content, 'html.parser')

# _4ddWXP --> for each product on a page, _2rpwqI --> each product's new Redirect URL
# _1fQZEK --> for air conditioners
# s1Q9rs --> for headphones one ear
# below the class needs to be changed as per The Flipkart Layout

products = soup.find_all('a', {'class': "_1fQZEK"})
href_list = [link['href'] for link in products]
print(href_list)
```

```python
def get_data(req):
    review = []
    rating = []
    
    soup = bs(req.content, 'html.parser')
    ratings = soup.find_all('div', attrs={'class': lambda e: e.startswith('_3LWZlK') if e else False})
    revs = soup.find_all('div', {"class": "t-ZTKy"})
    
    for n in ratings[1:]: 
        rating.append(n.text)
    for rev in revs:
        review.append(rev.text)
    print(len(rating), len(review))
    return rating, review
```

```python
len(href_list)
```

```python
for i in range(len(href_list)):
        # Product URL create
        PRODUCT_URL = urljoin(BASE_URL, href_list[i])
        product_req = requests.get(PRODUCT_URL)
        product_soup = bs(product_req.content, 'html.parser')
        
        # detail of product --> title & description
        product_title = product_soup.find('span', {'class': 'B_NuCI'}).text
        product_desc = product_soup.find('div', {'class': "_1mXcCf RmoJUa"})
        if product_desc is None:
            product_desc = "none"
        else:
            product_desc = product_desc.text
            
        # extracting Review Page URL for the Product
        is_reviewBtn_present = product_soup.find('div', {'class': '_3UAT2v _16PBlm'})
        if is_reviewBtn_present is None: 
            ## reviews present on same page
            pass
        else:
            ## we have separate pages for reviews extract the URL for it
            review_page_param = product_soup.find('div', {'class': 'col JOpGWq'}).find('a')['href']
            review_page_param = review_page_param.split('FLIPKART')[0]+"FLIPKART"
            review_page_param = review_page_param.split('&')
            if len(review_page_param) >= 2:
                review_page_param.pop(-2)
            review_page_param = '&'.join(review_page_param)

            PRODUCT_REVIEW_URL = urljoin(BASE_URL, review_page_param)
            PRODUCT_REVIEW_URL = PRODUCT_REVIEW_URL+"&page=1"
            
            # does that page consist of a NEXT BUTTON 
            review_page_req = requests.get(PRODUCT_REVIEW_URL)
            review_page_soup = bs(review_page_req.content, 'html.parser')
            next_btn = review_page_soup.find('a', {"class": "_1LKTO3"})

            # extracting only 1000 reviews per PRODUCT
            j = 1
            while next_btn:
                ratings, reviews = get_data(review_page_req)

                new_data = pd.DataFrame({'title': product_title, 'description': product_desc,'reviews': reviews, 'ratings': ratings})
                df = pd.concat([df, new_data], ignore_index=True)
                
                print(f'{j}. {PRODUCT_REVIEW_URL}\n')
                PRODUCT_REVIEW_URL = urljoin(BASE_URL, next_btn['href'])
                review_page_req = requests.get(PRODUCT_REVIEW_URL)
                review_page_soup = bs(review_page_req.content, 'html.parser')
                next_btn = review_page_soup.find_all('a', {"class": "_1LKTO3"})
                
                if(len(next_btn) < 2):
                    print(next_btn)
                    print('\n')
                    break
                next_btn = next_btn[1]

                
                if j%10==0:
                    print(j)

                j = j+1
                if j > 100:
                    break

                if not next_btn:
                    break
        print(f'YES -- {i+1}')
```
